```python
import matplotlib.pyplot as plt
import astropy.units as u
from astropy.coordinates import EarthLocation, SkyCoord
from pytz import timezone
from astropy.time import Time
from astroplan import Observer
from astroplan import FixedTarget
from astroplan.plots import plot_airmass

## TODO charger le fichier *.information.csv du système dans Pandas systeme_df

#
## TODO prendre ces valeurs dans astrodm/observatoires.csv
longitude = '-73d47m35.4s'
latitude = '+45d35m42.3s'
elevation = 45 * u.m
location = EarthLocation.from_geodetic(longitude, latitude, elevation)

#
## TODO les valeurs d'environnement pourraient provenir d'un fichier csv
#  situé dans astrodm/environnements, avec une série de valeurs pour chaque saison
#
observer = Observer(name='Laval',
               location=location,
               pressure=0.615 * u.bar,
               relative_humidity=0.50,
               temperature=0 * u.deg_C,
               timezone=timezone('US/Eastern'),
               description="Site personnel, Laval, Québec, Canada")
#
#  TODO prendre les valeurs suivantes nécessaires dans systeme_df
#
coord = '05 14 32.27s -08 12 05.9s'
#coord = '05h14m32.27s -08d12m05.9s'
coordinates = SkyCoord('05h14m32.27s', '-08d12m05.9s', frame='icrs')
cible = FixedTarget(name='STF668', coord=coordinates)
```

```python
# time UTC???
# UTC observe_time = Time('2021-01-12 23:00:00')
observe_time = Time('2022-02-14 04:00:00')

plot_airmass(cible, observer, observe_time)
plt.tight_layout()
plt.show()
```
