## Instrumentation f/20 en «Lucky Imaging»

L'instrumentation pour le «Lucky Imaging» sur EdgeHD 8 est : ASI 120MM-S + Barlow 2X = f/20 appr. Le champ de la caméra est alors de 160 X 120 arcsec environ.

## Discussion sur les données du catalogue WDS : paire AB
D'après les données de catalogue ci haut, la paire AB avec sep2 = 57.8 arcsec, semble parfaite pour l'instrumentation f/20 et sera observée en premier. À la rigueur, la paire BD (sep2 = 133.70 arcsec) et la paire AC (sep2 = 155.7 arcsec) pouraient tout juste être observées à f/20. Par contre, la paire AD (sep2 = 181.4 arcsec) est trop espacée pour cette instrumentation, mais pourrait éventuellement être mesurée sans la Barlow donc à f/10. Cependant l'exploitation de l'instrumentation à f/10 reste à développer.
